```python
# Cs 1302 Group project by Alex Castro, Christy Lam, Hashim-Omar Omar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pokedex = pd.read_csv("Pokemon_full.csv", sep=",")
pokedex
```

```python
'''
since the data does not contain any Nan, we do not need to do any cleaning
'''
# 3. (20 points) Calculate some statistics.
# 3.1. Choose at least two features (i.e., columns) to calculate their mean, median, variance, standard deviation, etc. 
# 3.2. Describe what you find given the statistics.

# Separating the dataframe into each generation of pokemon, for our data we will focus on the first 4 generation(0 to 493)
# Gen 1 is from 0 to 151
# Gen 2 is from 152 to 251
# Gen 3 is from 252 to 386
# Gen 4 is from 387 to 493

gen1 = pokedex.loc[np.arange(0,151)]
gen2 = pokedex.loc[np.arange(151, 251)]
gen3 = pokedex.loc[np.arange(251, 386)]
gen4 = pokedex.loc[np.arange(386, 493)]
```

```python
# 3.1. Choose at least two features (i.e., columns) to calculate their mean, median, variance, standard deviation, etc. 
'''
We will use three columns for this section, the hp, attack, and defense stats of the pokemon: 

pokedex[['hp', 'attack', 'defense]]
'''
# The mean of the first 4 generation of pokemon

print("The mean of the Generation 1 is:")
print(gen1[['hp', 'attack', 'defense']].mean())

print("\nThe mean of Generation 2 is: ")
print(gen2[['hp', 'attack', 'defense']].mean())

print("\nThe mean of Generation 3 is: ")
print(gen3[['hp', 'attack', 'defense']].mean())

print("\nThe mean of Generation 4 is: ")
print(gen4[['hp', 'attack', 'defense']].mean())
```

```python
# The median of the first 4 generation of pokemon

print("The median of the Generation 1 is:")
print(gen1[['hp', 'attack', 'defense']].median())

print("\nThe median of Generation 2 is: ")
print(gen2[['hp', 'attack', 'defense']].median())

print("\nThe median of Generation 3 is: ")
print(gen3[['hp', 'attack', 'defense']].median())

print("\nThe median of Generation 4 is: ")
print(gen4[['hp', 'attack', 'defense']].median())
```

```python
# The variance of the first 4 generation of pokemon

print("The variance of the Generation 1 is:")
print(gen1[['hp', 'attack', 'defense']].var())

print("\nThe variance of Generation 2 is: ")
print(gen2[['hp', 'attack', 'defense']].var())

print("\nThe variance of Generation 3 is: ")
print(gen3[['hp', 'attack', 'defense']].var())

print("\nThe variance of Generation 4 is: ")
print(gen4[['hp', 'attack', 'defense']].var())
```

```python
# The standard deviation of the first 4 generation of pokemon

print("The standard deviation of the Generation 1 is:")
print(gen1[['hp', 'attack', 'defense']].std())

print("\nThe standard deviation of Generation 2 is: ")
print(gen2[['hp', 'attack', 'defense']].std())

print("\nThe standard deviation of Generation 3 is: ")
print(gen3[['hp', 'attack', 'defense']].std())

print("\nThe standard deviation of Generation 4 is: ")
print(gen4[['hp', 'attack', 'defense']].std())
```

```python
# 3.2. Describe what you find given the statistics.
print("The mean, median, variance, and standard deviation calculation looks at the battle statistics of the pokemon from the first four generation.")
print("the battle statistics of the pokemon incude the 'hp', 'attack, and 'defense' column")
print("The mean calculation shows Generation four had the highest mean value among the other 3 while Generation 1 had the lowest")
print("For median, Generation 4 also stood out having the highest values while Generation 3 had the lowest")
print("For the Variance calculations, Generation 2 rose against the rest, while Generation 4 had the lowest")
print("The final calculations was the standard deviation, which showed Generation 2 rising once again, while Generation 1 had the lowest")
print("Overall, The statistics showed that Generation 4 and 2 show greater potential in battle statistics compared to Gen 1 and 3")
```

```python
# Line plot
'''
The line plot will graph the speed of each pokemon from its generation
'''

fig, ax = plt.subplots(2, 2, figsize = (18,12))

plt.suptitle("Speed comparison of each generation")
ax[0, 0].plot(gen1['pokedex id'], gen1['speed'], 'b', label="Generation 1")
ax[0, 1].plot(gen2['pokedex id'], gen2['speed'], 'r', label="Generation 2")
ax[1, 0].plot(gen3['pokedex id'], gen3['speed'], 'y', label="Generation 3")
ax[1, 1].plot(gen4['pokedex id'], gen4['speed'], 'm', label="Generation 4")

ax[0, 0].set_title('Generation 1')
ax[0, 1].set_title('Generation 2')
ax[1, 0].set_title('Generation 3')
ax[1, 1].set_title('Generation 4')

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()
ax[1, 1].legend()

plt.show()
```

```python
# Bar Graph
data = pd.read_csv("Pokemon_full.csv").drop_duplicates(subset=['type'], keep="first")
poke_type = np.arange(len(data))
x_hp = data['hp']
x_attack = data['attack']

y_ticks = data['type'].tolist()
y_ticks = plt.yticks(np.arange(len(y_ticks)), y_ticks)

plt.barh(poke_type - 0.2, x_hp, 0.4, label = 'hp')
plt.barh(poke_type + 0.2, x_attack, 0.4, label = 'attack')

plt.xlabel('hp/attack')
plt.ylabel('type')
plt.title('type in comparision to hp/attack')
plt.legend()
plt.show()
```

```python
# histogram
fig, axs = plt.subplots(2, 2, figsize=(10, 12))


axs[0, 0].hist(gen1['weight'], color='r', label = 'Generation 1') 
axs[0, 0].set_title('Weight compared to Gen 1')
axs[0, 0].set_xlabel('Weight')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].legend()

axs[0, 1].hist(gen2['weight'], color='b', label = 'Generation 2')
axs[0, 1].set_title('Weight compared to Gen 2')
axs[0, 1].set_xlabel('Weight')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].legend()

axs[1, 0].hist(gen3['weight'], color='y', label = 'Generation 3')
axs[1, 0].set_title('Weight compared to Gen 3')
axs[1, 0].set_xlabel('Weight')
axs[1, 0].set_ylabel('Frequency')
axs[1, 0].legend()

axs[1, 1].hist(gen4['weight'], color='g', label = 'Generation 4')
axs[1, 1].set_title('Weight compared to Gen 4')
axs[1, 1].set_xlabel('Weight')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend()

# Add a main title for the figure
fig.suptitle('Weight Distribution for Each Generation')

# Show the plot
plt.show()
```

```python
# Pie Chart

# the dictionaries needed which convert the type names to numbers
gen1_type_dict = {
    'grass': 1,
    'fire': 2,
    'water': 3,
    'bug': 4,
    'normal': 5,
    'poison': 6,
    'electric': 7,
    'ground': 8,
    'fairy': 9,
    'fighting': 10,
    'psychic': 11,
    'rock': 12,
    'ghost': 13,
    'ice': 14,
    'dragon': 15
}

gen2_type_dict = {
    'grass': 1,
    'fire': 2,
    'water': 3,
    'bug': 4,
    'normal': 5,
    'poison': 6,
    'electric': 7,
    'ground': 8,
    'fairy': 9,
    'fighting': 10,
    'psychic': 11,
    'rock': 12,
    'ghost': 13,
    'ice': 14,
    'dark': 15,
    'steel': 16
}


gen3_type_dict = {
    'grass': 1,
    'fire': 2,
    'water': 3,
    'bug': 4,
    'normal': 5,
    'poison': 6,
    'electric': 7,
    'ground': 8,
    'fighting': 9,
    'psychic': 10,
    'rock': 11,
    'ghost': 12,
    'ice': 13,
    'dragon': 14,
    'dark': 15,
    'steel': 16
}

gen4_type_dict = {
    'grass': 1,
    'fire': 2,
    'water': 3,
    'bug': 4,
    'normal': 5,
    'poison': 6,
    'electric': 7,
    'ground': 8,
    'fairy': 9,
    'fighting': 10,
    'psychic': 11,
    'rock': 12,
    'ghost': 13,
    'ice': 14,
    'dragon': 15,
    'dark': 16,
    'steel': 17
}


gen1['type_num'] = gen1['type'].map(gen1_type_dict)
gen2['gen2_type_num'] = gen2['type'].map(gen2_type_dict)
gen3['gen3_type_num'] = gen3['type'].map(gen3_type_dict)
gen4['gen4_type_num'] = gen4['type'].map(gen4_type_dict)


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(21, 16))

axes[0, 0].pie(gen1['type_num'].value_counts(), labels=list(gen1_type_dict.keys()), autopct='%1.1f%%')
axes[0, 0].set_title('Gen 1')

axes[0, 1].pie(gen2['gen2_type_num'].value_counts(), labels=list(gen2_type_dict.keys()), autopct='%1.1f%%')
axes[0, 1].set_title('Gen 2')

axes[1, 0].pie(gen3['gen3_type_num'].value_counts(), labels=list(gen3_type_dict.keys()), autopct='%1.1f%%')
axes[1, 0].set_title('Gen 3')

axes[1, 1].pie(gen4['gen4_type_num'].value_counts(), labels=list(gen4_type_dict.keys()), autopct='%1.1f%%')
axes[1, 1].set_title('Gen 4')


fig.suptitle('Type Distribution for the 4 Generations', fontsize=16)

plt.show()
```
